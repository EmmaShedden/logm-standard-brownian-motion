a): We start with a one-dimensional motion. Draw a coordinate system with time 
t on the horizontal axis, and height h on the vertical axis. Mark the origin. Now, flip a coin. If heads, mark a point one step ahead and one step above the previously marked point. If tails, mark a point one step ahead and one step below the previous one. Continue this for a while and draw the resulting graph.

```python
import numpy as np
import matplotlib.pyplot as plt

myarray = np.random.randint(0,2,10000)
#print(myarray)
```

```python
xs = np.array([0])
ys = np.array([0])
average_h = np.array([0])
average_h2 = np.array([0])
#count0 = 0
#count1 = 0

for i in range(len(myarray)):
    xs = np.append(xs,i+1)
    if myarray[i] == 0:
        ys = np.append(ys,ys[i]+1)
        average_h = np.append(average_h,sum(ys)/(i+2))
        average_h2 = np.append(average_h2,sum(ys**2)/(i+2))
        #count0 += 1
    elif myarray[i] == 1:
        ys = np.append(ys,ys[i]-1)
        average_h = np.append(average_h,sum(ys)/(i+2))
        average_h2 = np.append(average_h2,sum(ys**2)/(i+2))
        #count1 += 1

xmin, xmax, ymin, ymax = 0, 10000, -1000, 1000
ticks_frequency = 1

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(xs, ys,s = 0.5,c='#0F4C81')

plt.text(0.08,0.1,'(0,0)')

for i in range(1,len(xs)):
    ax.plot([xs[i], xs[i-1]], [ys[i], ys[i-1]], c='#0F4C81', ls='-', lw=0.5, alpha=0.5)
    
ax.set_xlabel('t', size=14, labelpad=-24, x=1.03)
ax.set_ylabel('h', size=14, labelpad=-21, y=1.02, rotation=0)

plt.show()

#print(count0)
#print(count1)
```

e): Next, you will draw a two-dimensional random walk. Draw a chessboard pattern around the origin, and roll a die. If you get a 1, mark the square to the left of the previous square. If 2, mark the square below the previous one. If 3, mark the one to the right, and if 4, mark the one above. If you get 5 or 6, roll again.

```python
pathx = np.array([0])
pathy = np.array([0])

for i in range(10000):
    die = np.random.randint(1,7,1)
    if die[0] == 1:
#        print(f'1 here {i}')
        pathx = np.append(pathx,pathx[i]-1)
        pathy = np.append(pathy,pathy[i])
    elif die[0] == 2:
#        print(f'2 here {i}')
        pathx = np.append(pathx,pathx[i])
        pathy = np.append(pathy,pathy[i]-1)
    elif die[0] == 3:
#        print(f'3 here {i}')
        pathx = np.append(pathx,pathx[i]+1)
        pathy = np.append(pathy,pathy[i])
    elif die[0] == 4:
#        print(f'4 here {i}')
        pathx = np.append(pathx,pathx[i])
        pathy = np.append(pathy,pathy[i]+1)
    else:
        pathx = np.append(pathx,pathx[i])
        pathy = np.append(pathy,pathy[i])

xmin, xmax, ymin, ymax = -10000, 10000, -10000, 10000
ticks_frequency = 1

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(pathx, pathy,s = 0.5,c='#0F4C81')

plt.text(0.08,0.1,'(0,0)')

for i in range(1,len(pathx)):
    ax.plot([pathx[i], pathx[i-1]], [pathy[i], pathy[i-1]], c='#0F4C81', ls='-', lw=0.5, alpha=0.5)

plt.show()
```
